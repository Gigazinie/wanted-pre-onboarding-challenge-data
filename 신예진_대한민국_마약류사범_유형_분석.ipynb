   ```
    가. 양귀비: 양귀비과(科)의 파파베르 솜니페룸 엘(Papaver somniferum L.), 파파베르 세티게룸 디시(Papaver setigerum DC.) 또는 파파베르 브락테아툼(Papaver bracteatum)
    나. 아편: 양귀비의 액즙(液汁)이 응결(凝結)된 것과 이를 가공한 것. 다만, 의약품으로 가공한 것은 제외한다.
    다. 코카 잎[엽]: 코카 관목[(灌木): 에리드록시론속(屬)의 모든 식물을 말한다]의 잎. 다만, 엑고닌ㆍ코카인 및 엑고닌 알칼로이드 성분이 모두 제거된 잎은 제외한다.
    라. 양귀비, 아편 또는 코카 잎에서 추출되는 모든 알카로이드 및 그와 동일한 화학적 합성품으로서 대통령령으로 정하는 것
    마. 가목부터 라목까지에 규정된 것 외에 그와 동일하게 남용되거나 해독(害毒) 작용을 일으킬 우려가 있는 화학적 합성품으로서 대통령령으로 정하는 것
    바. 가목부터 마목까지에 열거된 것을 함유하는 혼합물질 또는 혼합제제. 다만, 다른 약물이나 물질과 혼합되어 가목부터 마목까지에 열거된 것으로 다시 제조하거나 제제(製劑)할 수 없고, 그것에 의하여 신체적 또는 정신적 의존성을 일으키지 아니하는 것으로서 총리령으로 정하는 것[이하 “한외마약”(限外麻藥)이라 한다]은 제외한다.
   ```
  3. “**향정**신성의약품”이란 인간의 중추신경계에 작용하는 것으로서 이를 오용하거나 남용할 경우 인체에 심각한 위해가 있다고 인정되는 다음 각 목의 어느 하나에 해당하는 것으로서 대통령령으로 정하는 것을 말한다.
  ```
    가. 오용하거나 남용할 우려가 심하고 의료용으로 쓰이지 아니하며 안전성이 결여되어 있는 것으로서 이를 오용하거나 남용할 경우 심한 신체적 또는 정신적 의존성을 일으키는 약물 또는 이를 함유하는 물질
    나. 오용하거나 남용할 우려가 심하고 매우 제한된 의료용으로만 쓰이는 것으로서 이를 오용하거나 남용할 경우 심한 신체적 또는 정신적 의존성을 일으키는 약물 또는 이를 함유하는 물질
    다. 가목과 나목에 규정된 것보다 오용하거나 남용할 우려가 상대적으로 적고 의료용으로 쓰이는 것으로서 이를 오용하거나 남용할 경우 그리 심하지 아니한 신체적 의존성을 일으키거나 심한 정신적 의존성을 일으키는 약물 또는 이를 함유하는 물질
    라. 다목에 규정된 것보다 오용하거나 남용할 우려가 상대적으로 적고 의료용으로 쓰이는 것으로서 이를 오용하거나 남용할 경우 다목에 규정된 것보다 신체적 또는 정신적 의존성을 일으킬 우려가 적은 약물 또는 이를 함유하는 물질
    마. 가목부터 라목까지에 열거된 것을 함유하는 혼합물질 또는 혼합제제. 다만, 다른 약물 또는 물질과 혼합되어 가목부터 라목까지에 열거된 것으로 다시 제조하거나 제제할 수 없고, 그것에 의하여 신체적 또는 정신적 의존성을 일으키지 아니하는 것으로서 총리령으로 정하는 것은 제외한다.
  ```
  4. “**대마**”란 다음 각 목의 어느 하나에 해당하는 것을 말한다. 다만, 대마초[칸나비스 사티바 엘(Cannabis sativa L)을 말한다. 이하 같다]의 종자(種子)ㆍ뿌리 및 성숙한 대마초의 줄기와 그 제품은 제외한다.
  ```
    가. 대마초와 그 수지(樹脂)
    나. 대마초 또는 그 수지를 원료로 하여 제조된 모든 제품
    다. 가목 또는 나목에 규정된 것과 동일한 화학적 합성품으로서 대통령령으로 정하는 것
    라. 가목부터 다목까지에 규정된 것을 함유하는 혼합물질 또는 혼합제제
  ```

## **Ⅱ. 마약류범죄 발생 및 검거 현황**

````python
crime_df = pd.read_csv('/content/drive/MyDrive/인공지능응용개발자과정/06 미니프로젝트/범죄_발생_및_검거_현황_전국_2017-2021.csv', encoding='cp949')
````

````python
crime_df.head()
````

````python
crime_df0 = crime_df.T.reset_index()
crime_df0.columns = ['연도','죄종별','마약범죄']
crime_df0 = crime_df0.drop(0, axis=0)
crime_df0['연도'] = crime_df0['연도'].str.split(".", expand=True)[0]
crime_df0['마약범죄'] = crime_df0['마약범죄'].astype(float).round().astype(int)
````

````python
crime_df0
````

### **1. 마약류범죄 발생건수(건) 추이**

````python
crime_df1 = crime_df0[crime_df0['죄종별'].isin(['발생건수 (건)'])]
crime_df1
````

````python
# 시각화 lineplot
sns.lineplot(data=crime_df1, x='연도', y='마약범죄')
plt.title('[마약류범죄 발생건수(건) 추이]')
plt.ylim([6000,10000])
plt.grid(True)
plt.show()
````

### **2. 마약류범죄 검거건수(건) 추이**

````python
crime_df2 = crime_df0[crime_df0['죄종별'].isin(['검거건수 (건)'])]
crime_df2
````

````python
# 시각화 lineplot
sns.lineplot(data=crime_df2, x='연도', y='마약범죄', color='orange')
plt.title('[마약류범죄 검거건수(건) 추이]')
plt.grid(True)
plt.show()
````

### **3. 마약류범죄 검거인원(명) 추이**

````python
crime_df3 = crime_df0[crime_df0['죄종별'].isin(['검거인원 (명)'])]
crime_df3
````

````python
# 시각화 lineplot
sns.lineplot(data=crime_df3, x='연도', y='마약범죄', color='green')
plt.title('[마약류범죄 검거인원(명) 추이]')
plt.grid(True)
plt.show()
````

### **4. 소결**


````python
crime_df4 = pd.concat([crime_df1, crime_df2, crime_df3])
crime_df4
````

````python
# 전체 시각화 lineplot
sns.lineplot(data=crime_df4, x='연도',	y='마약범죄', hue='죄종별')
plt.legend()
plt.title('[마약류범죄 발생건수, 검거건수, 검거인원 추이 비교]')
plt.grid(True)
plt.show()
````

- 2017년부터 2021년까지 발생건수는 증가 및 감소를 반복하였으나 전체적으로 증가 추세에 있다.
- 2018년에 전년 대비 발생건수와 검거인원이 감소하기는 하였으나 전반적으로는 빠른 증가 추세에 있음을 확인할 수 있다.
- 발생건수와 검거건수는 연도별로 비슷한 추세를 보이고 있다. 마약류범죄 검거율은 높다고 할 수 있다.
- 검거인원 수는 발생건수보다 많게 나타나는데, 이는 한 건의 범죄에 여러 명의 범인이 연루될 수 있기 때문이다.

2. 마약류 사용예방을 위한 정책대상의 연령대를 기존보다 더욱 낮추고 또 각 연령대층에 맞는 적절한 예방교육의 강화가 필요하다. 특히 마약류사범의 미성년 연령대층이 점차 빠른 속도로 증가해오고 있는 사실을 고려하면, 정책 대상 연령대는 단순히 20대뿐만이 아닌 아동·청소년인 10대 미성년까지 그 대상을 확대하여야 할 것이다.
3. 마약류범죄 재범률 감소를 위해 마약류사범에 대한 치료재활 프로그램을 더욱 개선하고 형사절차에서 의무화되어야 한다. 기존 재활프로그램을 제대로 진단하여 질적으로 개선하는 노력이 필요하다. 지역사회 차원에서도 교육프로그램 또는 재활프로그램 집행기관과 참여 기회를 확대하여야 할 것이다.
